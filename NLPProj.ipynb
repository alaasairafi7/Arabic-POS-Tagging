```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk as nltk
```

```python
book_data = pd.read_csv("/content/Moanath.txt",sep='\t', header= None)
book_data.columns= ['body_text']
book_data.head(10)
```

```python
import string
import re
def remove_punctuation(text):
     #Remove punctuation marks
     text = re.sub(r'[!"\$%&\'()*+,\-.\/:;=#@?\[\\\]^_`{|}~،ًُ]','',text)
     return text
book_data['body_text_nopunc'] = book_data['body_text'].apply(lambda x : remove_punctuation(x))
book_data.head(10)
```

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt')
book_data ['Tokenized'] =book_data ['body_text_nopunc'].apply(lambda x: word_tokenize(x))
book_data.head(10)
```

```python
import string
import re
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.tag import RegexpTagger
nltk.download('punkt')

file_path = '/content/Kalimat.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    arabic_words = file.read().split()

arabic_words = [re.sub(r'[' + string.punctuation + '،ًُ]+', '', word) for word in arabic_words]

#Define regular expression patterns and corresponding tags
def create_pattern(tag_words, tag):
    pattern = [(r'\b(?:ب|و|ف|(?:ال|وال|فال|بال))?(' + '|'.join(tag_words) + r')\b', tag)]
    return pattern

pattern = create_pattern(arabic_words, 'جمع مؤنث سالم')

def posss(tokenzz):
    tagging = RegexpTagger(pattern)# Create a RegexpTagger with the defined patterns
    tags = tagging.tag(tokenzz)# Tag the tokens using the RegexpTagger
    return tags

book_data['pos'] = book_data['Tokenized'].apply(lambda x: posss(x))
book_data.head(10)
```

```python
def length(tags):
    arr = []
    for word, tag in tags:
        if tag == "جمع مؤنث سالم":
            arr.append(len(word))
    return arr
book_data['len'] = book_data['pos'].apply(lambda x: length(x))
book_data.head(20)
```

```python
def end(tags):
    arr2 = []
    for word, tag in tags:
        if tag == "جمع مؤنث سالم":
            if (word[-2:] == "ات"):
              arr2.append("1")
            else:
              arr2.append("0")

    return arr2
book_data['end'] = book_data['pos'].apply(lambda x: end(x))
book_data.head(50)
```
